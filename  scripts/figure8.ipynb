```python
%matplotlib inline
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "serif"
plt.rcParams.update({'mathtext.default':  'regular' })
```

```python
import matplotlib.pyplot as plt

wsf = [0.29,0.36,0.5,0.3]


with open("../EWF/marconi.txt", "r") as file:
    marconi_ewf= [float(line.strip()) for line in file if line.strip()]
with open("../WUE/marconi.txt", "r") as file:
    marconi_wue= [float(line.strip()) for line in file if line.strip()]
    
with open("../EWF/fugaku.txt", "r") as file:
    fugaku_ewf= [float(line.strip()) for line in file if line.strip()]
with open("../WUE/fugaku.txt", "r") as file:
    fugaku_wue= [float(line.strip()) for line in file if line.strip()]
    
with open("../EWF/polaris.txt", "r") as file:
    polaris_ewf= [float(line.strip()) for line in file if line.strip()]
with open("../WUE/polaris.txt", "r") as file:
    polaris_wue= [float(line.strip()) for line in file if line.strip()]

with open("../EWF/frontier.txt", "r") as file:
    frontier_ewf= [float(line.strip()) for line in file if line.strip()]
with open("../WUE/frontier.txt", "r") as file:
    frontier_wue= [float(line.strip()) for line in file if line.strip()]
    
wi_chicago = np.mean(polaris_wue)+np.mean(polaris_ewf)*1.65
wi_oak = np.mean(frontier_wue)+np.mean(frontier_ewf)*1.05
wi_kobe = np.mean(fugaku_wue)+np.mean(fugaku_ewf)*1.4
wi_bologna = np.mean(marconi_wue)+np.mean(marconi_ewf)*1.25

FontSize=13
wi=[wi_bologna,wi_kobe,wi_chicago,wi_oak]
```

```python


fig, axs = plt.subplots(nrows=1, ncols=3, gridspec_kw={'hspace': 0.4, 'wspace': 0.5, 'bottom': 0.2, 
                    'top': 0.8, 'right':0.995, 'left':0.17}, figsize=(8,3))

# first
axs[0].grid(axis='y', linestyle='--')
axs[0].tick_params(labelsize=FontSize)
axs[0].bar(np.arange(4), wi, width = 0.6, color = '#69DDFF', edgecolor = 'black',capsize=FontSize, label='Offsite Water Footprint',zorder=3)
axs[0].set_xticks([k for k in range(4)])
axs[0].set_ylim([0,12])
axs[0].set_yticks(range(0, 13, 4))
axs[0].tick_params(axis='both', which='major', labelsize=FontSize)
axs[0].set_ylabel("Water Intensity \n(L/kWh)", fontsize=FontSize, labelpad=-3)
axs[0].set_xticklabels(["Marconi","Fugaku",'Polaris','Frontier'], rotation=25, fontsize = 13) 


#second
axs[1].grid(axis='y', linestyle='--')
axs[1].tick_params(labelsize=FontSize)
axs[1].bar(np.arange(4), wsf, width = 0.6, color = '#006989', edgecolor = 'black',capsize=FontSize, label='Offsite Water Footprint',zorder=3)
axs[1].set_xticks([k for k in range(4)])
axs[1].set_ylim([0,0.6])
axs[1].set_yticks(np.arange(0, 0.7, 0.2))
axs[1].tick_params(axis='both', which='major', labelsize=FontSize)
axs[1].set_ylabel("Water Scarcity Index", fontsize=FontSize, labelpad=3)
axs[1].set_xticklabels(["Marconi","Fugaku",'Polaris','Frontier'], rotation=25, fontsize = 13) 


#third
# first approach
axs[2].grid(axis='y', linestyle='--')
wi_wsf = [wi_bologna*(wsf[0]),wi_kobe*(wsf[1]),wi_chicago*(wsf[2]),wi_oak*(wsf[3])]
axs[2].tick_params(labelsize=FontSize)
axs[2].bar(np.arange(4), wi_wsf, width = 0.6, color = '#69DDFF', edgecolor = 'black',capsize=FontSize, label='Offsite Water Footprint',zorder=3,hatch='//')
axs[2].set_xticks([k for k in range(4)])
axs[2].set_ylim([0,4])
axs[2].set_yticks(np.arange(0, 5, 1))
axs[2].tick_params(axis='both', which='major', labelsize=FontSize)
axs[2].set_ylabel("Adjusted Water \nIntensity (L/kWh)", fontsize=FontSize, labelpad=1)
axs[2].set_xticklabels(["IT-NO","JP",'PJM','TVA'], rotation=0, fontsize = FontSize) 
axs[2].set_xticklabels(["Marconi","Fugaku",'Polaris','Frontier'], rotation=25, fontsize = 13) 
axs[0].text(0.5, -0.35, '(a)', fontsize=FontSize, ha='center', va='center', transform=axs[0].transAxes)
axs[1].text(0.5, -0.35, '(b)', fontsize=FontSize, ha='center', va='center', transform=axs[1].transAxes)
axs[2].text(0.5, -0.35, '(c)', fontsize=FontSize, ha='center', va='center', transform=axs[2].transAxes)

plt.savefig("../figures/figure8.pdf",bbox_inches='tight')

```
